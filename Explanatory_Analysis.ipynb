## Investigation Overview:
### This dataset presents the loan information for Prosper Personal Loans. 
#### Background on Prosper Loans:
**Based on the Prosper website, Prosper started in 2005 as a lender for personal loans through a peer-to-peer process. According to Forbes.com, 'Prosper is a peer-to-peer lending platform that lets borrowers access personal loans between 2,000 to 50,000 dollars. Borrower qualifications are more accessible than many other online lenders, but rates can reach almost 36% and borrowers pay origination fees from 1% to 9.99%... So Prosper Loans appear to be for people with decent or fair credit needing cash to pay off debts, school loans, or pay off purchases. The 'ListingCategory' variable from the dataset shows what types of personal loans are being financed in this dataset. I will check to see if the analysis from this dataset matches with Forbes' review of the peer to peer lender. Here are a couple of key terms that will be explained below from the dataset that will make the analysis clearer to understand.**

1. BorrowerAPR: is the borrower's annual percentage rate for the loan. The APR will be listed in decimals. So for example, 0.01315 will be 1.3%.
2. ListingCategory: The type of loan the borrower is obtaining.
3. ProsperScore: The custom built risk score from 1 to 10. 1 is the worst and 10 is the best.
4. LoanStatus: Reports if the loan is Cancelled, Charedoff, Completed, Current, Defaulted, FinalPaymentInProgress, or Past Due.
5. LoanOriginalAmount: The loan amount the borrower received from the peer to peer funding/lending process.
6. LoanOriginationDate: The date the loan was originated.
7. PercentFunded: The percentage of the loan was funded by peers (investors).
8. IncomeRange - The income range of the borrower at the time the listing was created.

## Dataset Overview:

**The dataset has 35 variables and 113,937 rows.**

**Below is the data dictionary for the Prosper Loan dataset:**

1. ListingNumber - The number that uniquely identifies the listing to the public as displayed on the website.
2. ListingCreationDate - The date the listing was created.
3. CreditGrade - The Credit rating that was assigned at the time the listing went live. Applicable for listings pre-2009 period and will only be populated for those listings.
4. Term	- The length of the loan expressed in months.
5. LoanStatus - The current status of the loan: Cancelled,  Chargedoff, Completed, Current, Defaulted, FinalPaymentInProgress, PastDue. The PastDue status will be accompanied by a delinquency bucket. 
6. BorrowerAPR - The Borrower's Annual Percentage Rate (APR) for the loan.
7. BorrowerRate	- The Borrower's interest rate for this loan. 
8. ProsperRating (numeric) - The Prosper Rating assigned at the time the listing was created: 0 - N/A, 1 - HR, 2 - E, 3 - D, 4 - C, 5 - B, 6 - A, 7 - AA.  Applicable for loans originated after July 2009.
9. ProsperRating (Alpha) - The Prosper Rating assigned at the time the listing was created between AA - HR.  Applicable for loans originated after July 2009.
10. ProsperScore - A custom risk score built using historical Prosper data. The score ranges from 1-10, with 10 being the best, or lowest risk score.  Applicable for loans originated after July 2009.
11. ListingCategory - numeric - The category of the listing that the borrower selected when posting their listing: 0 - Not Available, 1 - Debt Consolidation, 2 - Home Improvement, 3 - Business, 4 - Personal Loan, 5 - Student Use, 6 - Auto, 7- Other, 8 - Baby&Adoption, 9 - Boat, 10 - Cosmetic Procedure, 11 - Engagement Ring, 12 - Green Loans, 13 - Household Expenses, 14 - Large Purchases, 15 - Medical/Dental, 16 - Motorcycle, 17 - RV, 18 - Taxes, 19 - Vacation, 20 - Wedding Loans
12. BorrowerState - The two letter abbreviation of the state of the address of the borrower at the time the Listing was created.
13. Occupation - The Occupation selected by the Borrower at the time they created the listing.
14. EmploymentStatus - The employment status of the borrower at the time they posted the listing.
15. IsBorrowerHomeowner	- A Borrower will be classified as a homeowner if they have a mortgage on their credit profile or provide documentation confirming they are a homeowner. 
16. CreditScoreRangeLower - The lower value representing the range of the borrower's credit score as provided by a consumer credit rating agency.
17. CreditScoreRangeUpper - The upper value representing the range of the borrower's credit score as provided by a consumer credit rating agency. 
18. CurrentDelinquencies - Number of accounts delinquent at the time the credit profile was pulled.
19. AmountDelinquent - Dollars delinquent at the time the credit profile was pulled.
20. DebtToIncomeRatio -	The debt to income ratio of the borrower at the time the credit profile was pulled. This value is Null if the debt to income ratio is not available. This value is capped at 10.01 (any debt to income ratio larger than 1000% will be returned as 1001%).
21. IncomeRange - The income range of the borrower at the time the listing was created.
22. IncomeVerifiable - The borrower indicated they have the required documentation to support their income.
23. StatedMonthlyIncome	- The monthly income the borrower stated at the time the listing was created.
24. LoanNumber - Unique numeric value associated with the loan.
25. LoanOriginalAmount - The origination amount of the loan.
26. LoanOriginationDate - The date the loan was originated.
27. MonthlyLoanPayment - The scheduled monthly loan payment.
28. PercentFunded -	Percent the listing was funded.
29. Recommendations	- Number of recommendations the borrower had at the time the listing was created.
30. Investors - The number of investors that funded the loan.
31. ListingCategory - The listing category of loans without the leading numbers in front of them like the ListingCateogry - numeric had. 
32. OccupationFiltered - The occupations of the borrowers with all of the different types of students combined to just Student. 
33. DTI_Bin - The DebtToIncome Ratios grouped by the following: 'Low', 'Moderate', 'High', 'Very High', 'Critical'.
34. CreditRange - Credit scores of the borrowers that were grouped into: 'Poor credit', 'Fair credit', 'Good credit', 'Very good credit', 'Excellent credit'.
35. HighRisk -  Flags loans as high-risk based on specific conditions: CurrentDelinquencies >= 10, DebtToIncomeRatio >= 0.30, and CreditScoreRangeUpper <= 600.

## (Visualization 1):
**Examining the 'ListingCategory' in the states with the most Prosper borrowers.**

```python
# Creating a crosstab
crosstab = pd.crosstab(df_prosper_subset['ListingCategory'], df_prosper_subset['BorrowerState'])

# Sum counting per state and adding a sorting to show the top 10 states
top_10_states = crosstab.sum().sort_values(ascending=False).head(10).index

# Filter the crosstab to include only the top 10 states
crosstab_10 = crosstab[top_10_states]

# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(crosstab_10, annot=True, fmt='d', cmap='YlGnBu', linewidths=0.5)

# Adding the labels
plt.title('Heatmap of ListingCategory vs BorrowerState')
plt.xlabel('BorrowerState')
plt.xticks(rotation=45)
plt.ylabel('ListingCategory')

plt.show()
```

**So most borrowers are in California with a LoanCategory of Debt Consolidation. The Debt Consolidation number could possibly be even higher if there were Debt Consolidation loans listed under Not Available.**

## (Visualization 2)
**What variable helps the most to discern whether a BorrowerAPR is higher or lower.**

```python
# A Hexbin plot seeking if a higher ProsperScore means a Lower APR for the borrower.
df_prosper_subset.plot.hexbin(x='BorrowerAPR', y='ProsperScore', gridsize=30, cmap='Blues');
plt.title('Does Higher ProsperScore mean Lower APR?');
plt.show();
```

**Yes the hexbin chart shows that borrowers with a higher ProsperScore will generally pay a lower BorrowerAPR. The lower the ProsperScores are, the higher the BorrowerAPR will be. A ProsperScore of 8 had the highest densities for higher ProsperScore and ranged between 10% and 20% in BorrowerAPRs.**

```python
plt.figure(figsize=(10,6))
# Scatter plot with three variables
sns.scatterplot(data=df_prosper_subset, x='BorrowerAPR', y='LoanOriginalAmount', hue='ProsperScore', palette="Paired")
# Adding the title
plt.title('Borrower APRs vs Loan Original Amount')
plt.show()
```

**This Scatterplot clearly shows that there is a strong correlation between higher ProsperScores and lower BorrwerAPRs with all of the light Purples, dark Purples, and Beige colors populating that area. However, the loans with BorrowerAPRs of over 40% have ProsperScores of 4 and 5. For the loans between 30% and 40% BorrwerAPRs, the ProsperScore will vary greatly between a ProsperScore of 1 & 8 it appears. SO higher ProsperScores will correlate to lower Borrowers APRs, but higher BorrowerAPRs do not necessarily correlate to the lowest ProsperScores.**

**According to this chart, loans above 25,000, are have predominantly higher ProsperScores. There are only a few loans above 25,000 with lower ProsperScores noted with the Greenish dots (ProsperScore of 3 and 4).**

**Further analysis for what variable helps the most to discern whether a BorrowerAPR is higher or lower.**

```python
# Calculate correlation matrix
corr = df_prosper_subset[['BorrowerAPR', 'ProsperScore', 'DebtToIncomeRatio', 'CreditScoreRangeUpper', 'LoanOriginalAmount']].corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap for Key Variables')
plt.show()
```

**The BorrowerAPR vs ProsperScore: ProsperScore has the highest negative correlation at -0.67. This means as the ProsperScore goes up, the BorrowerAPR will generally decrease. However, I am not sure how Prosper's custom ProsperScore was calculated except that Prosper uses historical data.**

**So to answer question 2, borrowers with a higher ProsperScore will generally pay a lower BorrowerAPR.**

#### 5. What is the most important takeaway of this analysis?

```python
plt.figure(figsize=(12, 8))

# Scatter plot with multiple encodings
sns.scatterplot(data=df_prosper_subset, 
                x='BorrowerAPR', 
                y='LoanOriginalAmount', 
                hue='CreditRange', 
                size='PercentFunded', 
                style='Term', 
                palette='coolwarm', sizes=(20, 200))

# Adding title, labels, and a legend
plt.title('BorrowerAPR vs LoanOriginalAmount with CreditRange, PercentFunded, and Term')
plt.xlabel('BorrowerAPR')
plt.ylabel('LoanOriginalAmount')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

**This multivariate scatterplot shows that the majority of loans with very good and excellent credit have lower BorrowerAPRs, while the majority of the loans with poor and fair credit had BorrowerAPRs above 20%. However, there were some loans that had poor, fair, and good credit as seen by the blue dots on the bottom left of the chart. I am not quite sure how these borrower's received such favorable BorrowerAPRs. I guess it has to do with these aforementioned loans being predominantly $5,000 or less.**
